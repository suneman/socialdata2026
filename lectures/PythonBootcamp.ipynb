## Part 3: Using LLMs for Coding

Large Language Models have transformed how we write code. In this course, we encourage you to use LLMs as coding assistants, but with intention.

### When LLMs are helpful
- **Boilerplate code**: Setting up plots, reading files, standard data transformations
- **Debugging**: "Why am I getting this error?" often gets good answers
- **Learning new libraries**: "How do I do X in pandas?" is faster than searching docs (although I feel ambivalent about this since it means great sites like stackoverflow are dying)
- **Code explanation**: Understanding unfamiliar code you encounter
- **Styling plots and making them look good**: If you can tell an LLM what you want a plot to look like, it can save lots of time with matplotlib documentation.

### When to be careful
- **Core concepts**: If you don't understand *why* something works, you'll struggle to adapt it
- **Analysis decisions**: LLMs don't know your data or research questions
- **Blindly copying**: Always read and understand code before using it
- **Complex logic**: LLMs can confidently produce subtly wrong code

### My expectation
You should be able to explain any code you submit. If you used an LLM to help write parts of it, that's fine, but make sure you understand what it does and why. The goal is learning, and LLMs are a tool to help you learn. The trouble is that usually it means that you avoid learning.

## Part 4: Pandas Essentials

Pandas is the backbone of much of data analysis in Python. Think of it as Excel with superpowers. Let's walk through the essentials you'll use constantly in this course.

For a more comprehensive introduction, see the official [10 Minutes to Pandas](https://pandas.pydata.org/pandas-docs/stable/user_guide/10min.html) tutorial.

```python
import pandas as pd
import numpy as np
```

### Creating DataFrames

A DataFrame is a 2D table with labeled rows and columns.

```python
# From a dictionary
df = pd.DataFrame({
    "city": ["Copenhagen", "Aarhus", "Odense", "Aalborg"],
    "population": [794128, 285273, 180302, 119862],
    "area_km2": [88.25, 91.00, 304.34, 139.00]
})

df
```

### Selecting Data

Use `.loc[]` for label-based selection and `.iloc[]` for position-based selection.

```python
# Select a single column (returns a Series)
df["city"]
```

```python
# Select multiple columns (returns a DataFrame)
df[["city", "population"]]
```

```python
# Select rows by condition
df[df["population"] > 200000]
```

```python
# Select specific rows and columns with .loc
df.loc[df["population"] > 200000, ["city", "population"]]
```

### Creating New Columns

```python
# Calculate population density
df["density"] = df["population"] / df["area_km2"]
df
```

### Basic Statistics

```python
# Quick summary statistics
df.describe()
```

```python
# Individual statistics
print(f"Total population: {df['population'].sum():,}")
print(f"Mean population: {df['population'].mean():,.0f}")
print(f"Max density: {df['density'].max():,.0f} people/km2")
```

## Part 5: Matplotlib Basics

Matplotlib is Python's foundational plotting library. While there are many visualization libraries built on top of it, understanding matplotlib gives you full control over your visualizations.

Full documentation: [matplotlib.org](https://matplotlib.org/)

```python
import matplotlib.pyplot as plt
```

### Basic Line Plot

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.title("A Simple Sine Wave")
plt.show()
```

### Bar Chart

```python
plt.figure(figsize=(8, 4))
plt.bar(df["city"], df["population"] / 1000)
plt.xlabel("City")
plt.ylabel("Population (thousands)")
plt.title("Population of Danish Cities")
plt.show()
```

### Scatter Plot

```python
plt.figure(figsize=(8, 4))
plt.scatter(df["area_km2"], df["population"], s=100)

# Add city labels
for i, city in enumerate(df["city"]):
    plt.annotate(city, (df["area_km2"].iloc[i], df["population"].iloc[i]), 
                 textcoords="offset points", xytext=(5, 5))

plt.xlabel("Area (km)")
plt.ylabel("Population")
plt.title("City Size vs Population")
plt.show()
```

### Subplots

Often you'll want multiple plots side by side.

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Left plot: population
axes[0].bar(df["city"], df["population"] / 1000, color="steelblue")
axes[0].set_ylabel("Population (thousands)")
axes[0].set_title("Population")

# Right plot: density
axes[1].bar(df["city"], df["density"], color="coral")
axes[1].set_ylabel("People per km")
axes[1].set_title("Population Density")

plt.tight_layout()
plt.show()
```
